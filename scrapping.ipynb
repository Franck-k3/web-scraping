```python
#importation des packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#saisir url
url="http://feeds.bbci.co.uk/news/rss.xml"
```

```python
#contenu dde la page
reponse=requests.get(url)
soup=BeautifulSoup(reponse.text,"html.parser")
soup
```

```python
items=soup.findAll('item')
```

```python
item=items[0]
item.title.text
```

```python
news_items=[]
```

```python
for i in items:
  news_i={}
  news_i['title']=i.title.text
  news_i['description']=i.description.text
  news_i['pubdate']=i.pubdate.text
  news_items.append(news_i)
```

```python
news_items
```

```python
#on veut une data base
journal=pd.DataFrame(news_items,columns=['title','description','pubdate'])
```
